Exercicio: implementar um codigo que EXECUTE as varias linhas de programa gerados pela conversao dos comandos para instrucoes de maquina. O resultado final deve aparecer na memoria ao final da execucao.

```python
#guilherme scher

memoria = ['0000','0000','0000','0000']
acumulador = ['0000']

def convert4Bits(parametro):
    tamanhoRes = len(parametro)
    if tamanhoRes < 4:
        for repeticoes in range(4-tamanhoRes):
            parametro = '0'+parametro
    if tamanhoRes > 4:
        parametro = parametro[-4:]
    return parametro

def entrada():
    n1 = int(input('Digite o primeiro valor (1 - 15): '))
    n2 = int(input('Digite o segundo valor (1 - 15): '))
    if n1 >= 1 and n1 <= 15 and n2 >= 1 and n2 <= 15:
        print(n1, 'x', n2)
        print('Valor aceito')
        return n1, n2
    else:
        print('Valor nao aceito')
        return -1, -1

def mostraMemoria():
    mem = ''
    for posicao in memoria:
        mem = mem+posicao+ ' '
    print('Memoria: ', mem)

def mostraAcumulador():
    print('AC:',acumulador)

def ler(posicao):
    global acumulador
    acumulador = memoria[posicao]
    mostraAcumulador()

def escrever(posicao, dado):
    global acumulador, memoria
    memoria[posicao] = dado
    mostraMemoria()

def somar(endDec, dado):
    global acumulador, memoria
    valor = memoria[endDec]
    valorDec = int(valor, 2)
    dadoDec = int(dado, 2)
    resultado = valorDec + dadoDec
    resultadoBin = bin(resultado)[2:]
    resConvertido = convert4Bits(resultadoBin)
    memoria[endDec] = resConvertido
    acumulador = resConvertido
    mostraMemoria()
    mostraAcumulador()

def subtrair(endDec, dado):
    global acumulador, memoria
    valor = memoria[endDec]
    valorDec = int(valor, 2)
    dadoDec = int(dado, 2)
    resultado = valorDec - dadoDec
    if resultado < 0:
        resultado = 0  # serve pra nao dar erro caso seja nÃºmero negativo :)
    resultadoBin = bin(resultado)[2:]
    resConvertido = convert4Bits(resultadoBin)
    memoria[endDec] = resConvertido
    acumulador = resConvertido
    mostraMemoria()
    mostraAcumulador()

def multiplicar(v1, v2):
    v2Bin = bin(v2)[2:]
    v2Mem = convert4Bits(v2Bin)
    escrever(0, v2Mem)

    for i in range(v1-1):
        somar(0, v2Mem)

def executaInstrucao(instr):
    com = instr[:2]
    comDec = int(com, 2)
    end = instr[2:4]
    endDec = int(end, 2)
    dado = instr[4:]
    
    if comDec == 0:
        ler(endDec)
    elif comDec == 1:
        escrever(endDec, dado)
    elif comDec == 2:
        somar(endDec, dado)
    elif comDec == 3:
        subtrair(endDec, dado)

v1, v2 = entrada()

if v2 != -1 and v1 != -1:
    multiplicar(v1, v2)
    mostraMemoria()
    mostraAcumulador()
```
